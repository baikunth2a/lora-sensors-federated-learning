```python
import pandas as pd
import time
import socket
import pickle
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras as K
from keras.models import load_model
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('dataset.csv', index_col='district_name')
X, y = data.iloc[:,:-5].to_numpy(), data.iloc[:,-5:].to_numpy()
#data.head()
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
y = scaler.fit_transform(y)
```

```python
model = Sequential()
model.add(Dense(2, input_dim=X.shape[1], activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(y.shape[1], activation='linear'))
model.compile(loss='mean_squared_error', optimizer=Adam(learning_rate=0.01))
```

```python
def receive_model():
    data = b''
    while True:
        packet = client.recv(4096)
        data += packet
        if b'END' in data:
            with open('models/model_c3.h5', 'wb') as f:
                f.write(data[:data.index(b'END')])
            break
```

```python
def train_model():
    model = K.models.load_model('models/model_c3.h5')
    model.fit(X,y, epochs=10)
    model.save('models/model_c3.h5', save_format='h5')
```

```python
def send_model():
    with open('models/model_c3.h5', 'rb') as f:
            data = f.read()
            client.sendall(data)
            client.sendall(b'END')
```

```python
client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
client.connect(('localhost', 12345))
time.sleep(0.5)
```

```python
i = 0

receive_model()
client.sendall("model_initialized".encode())

while True:
    client.sendall("request_global_model".encode())
    receive_model()
    train_model()
    
    message = client.recv(4096).decode()
    if message == "send_updated_model":
        print("Model requested!")
        send_model()
    else:
        print("Invalid message received from server.")
    print(f"Model train cycle {i}")
    time.sleep(2)
    i = i+1
client.shutdown(socket.SHUT_WR)
client.close()
```
